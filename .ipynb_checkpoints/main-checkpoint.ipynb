```python
import pygame
import math
import heapq
import random


# Input rows and cols

while True:
    rows=input('Enter number of rows (10-60):')
    try:
        rows = int(rows)
        if 10 <= rows <= 60:
            break
        else:
            print('Number of rows out of limit!')
    except ValueError:
        print('Invalid Input!')

while True:
    cols = input('Enter number of columns (10-60):')
    try:
        cols = int(cols)
        if 10 <= cols <= 60:
            break
        else:
            print('Number of columns out of limit!')
    except ValueError:
        print('Invalid Input!')

while True:
    density = input('Enter initial random wall density (0.0 - 0.6):')
    try:
        density = float(density)
        if 0.0 <= density <= 0.6:
            break
        else:
            print('Density out of range!')
    except ValueError:
        print('Invalid Input!')

#Constants

EMPTY = 0
WALL = 1
START = 2
GOAL = 3
PATH = 4
VISITED = 5
FRONTIER = 6
AGENT = 7

WHITE = (255,255,255)
BLACK = (0,0,0)
GRAY = (100,100,100)
GREEN = (0,200,0)
RED = (200,0,0)
BLUE = (0,0,255)
YELLOW = (255,255,0)
PURPLE = (150,0,200)

GRID_WIDTH = 600
PANEL_WIDTH = 250
WINDOW_HEIGHT = 800

#Initialize Pygame

pygame.init()
cellSize = GRID_WIDTH // max(rows, cols)
screen = pygame.display.set_mode((GRID_WIDTH + PANEL_WIDTH, WINDOW_HEIGHT))
pygame.display.set_caption("Dynamic Pathfinding Agent")


#Button Class

class Button:
    def __init__(self, x,y,width,height,text):
        self.rect = pygame.Rect(x,y,width,height)
        self.text = text
        self.selected = False

    def draw(self):
        color = (180, 180, 180) if self.selected else (220, 220, 220)
        pygame.draw.rect(screen, color, self.rect)
        pygame.draw.rect(screen, BLACK, self.rect, 2)
        font = pygame.font.SysFont(None, 24)
        t = font.render(self.text, True, BLACK)
        screen.blit(t,(self.rect.x+10,self.rect.y+10))

    def clicked(self, position):
        return self.rect.collidepoint(position)


#Buttons

manhattanBtn = Button(620,50,200,40,"Manhattan")
euclideanBtn = Button(620,100,200,40,"Euclidean")

astarBtn = Button(620,170,200,40,"A*")
gbfsBtn = Button(620,220,200,40,"GBFS")

generateBtn = Button(620,290,200,40,"Generate Map")

startBtn = Button(620,360,200,40,"Start")
pauseBtn = Button(620,410,200,40,"Pause")
stopBtn = Button(620,460,200,40,"Stop")

dynamicBtn = Button(620, 510, 200, 40, "Dynamic: OFF")

#Global Variables

grid = [[EMPTY for _ in range(cols)] for _ in range(rows)]

startPos = None
goalPos = None
agentPos = None
agentPath = []
agentIndex = 0

mode = "wall"
heuristicType = "manhattan"
algorithmType = "astar"

algorithmRunning = False
algorithmPaused = False
agentMoving = False

frontier = []
parent = {}
gValue = {}

nodesVisited = 0
pathCost = 0
startTime = 0
executionTime = 0

dynamicMode = False
spawnProbability = 0.5


# HEURISTICS


def manhattan(a,b):
    return abs(a[0]-b[0]) + abs(a[1]-b[1])

def euclidean(a,b):
    return math.sqrt((a[0]-b[0])**2 + (a[1]-b[1])**2)

def heuristic(a,b):
    if heuristicType=="manhattan":
        return manhattan(a,b)
    return euclidean(a,b)

#Map Generation

def generate_random_map():
    global grid
    for row in range(rows):
        for col in range(cols):
            if random.random() < density:
                grid[row][col] = WALL
            else:
                grid[row][col] = EMPTY

#Clear Screen

def clear_search():
    global grid
    for row in range(rows):
        for col in range(cols):
            if grid[row][col] in (PATH, VISITED, FRONTIER, AGENT):
                grid[row][col] = EMPTY
    if startPos:
        grid[startPos[0]][startPos[1]] = START
    if goalPos:
        grid[goalPos[0]][goalPos[1]] = GOAL


#Initialize Search

def init_search():
    global frontier, parent, gValue, nodesVisited, startTime
    frontier = []
    parent = {}
    gValue = {startPos:0}
    heapq.heappush(frontier, (0, startPos))
    if not agentMoving:
        nodesVisited = 0
    if not agentMoving:
        startTime = pygame.time.get_ticks()

#Search Step

def search_step():
    if not startPos or not goalPos:
        return False
    global frontier, nodesVisited, executionTime, pathCost

    if not frontier:
        return False

    current = heapq.heappop(frontier)[1]
    nodesVisited += 1

    if current == goalPos:
        executionTime = pygame.time.get_ticks() - startTime
        reconstruct_path(current)
        pathCost = len(agentPath)
        return False

    row, col = current
    if grid[row][col] not in (START, GOAL):
        grid[row][col] = VISITED

    for nextRow,nextCol in get_neighbors(current):
        if grid[nextRow][nextCol] == WALL:
            continue

        tempGValue = gValue[current]+1
        if (nextRow, nextCol) not in gValue or tempGValue < gValue[(nextRow, nextCol)]:
            parent[(nextRow, nextCol)] = current
            gValue[(nextRow, nextCol)] = tempGValue

            if algorithmType == "astar":
                fValue = tempGValue + heuristic((nextRow, nextCol),goalPos)
            else:
                fValue = heuristic((nextRow, nextCol), goalPos)

            heapq.heappush(frontier, (fValue, (nextRow, nextCol)))
            if grid[nextRow][nextCol] == EMPTY:
                grid[nextRow][nextCol] = FRONTIER
    return True

#Reconstruct path

def reconstruct_path(current):
    global agentPath, agentIndex, agentMoving
    path = []
    while current in parent:
        path.append(current)
        current = parent[current]
    path.reverse()
    agentPath = path
    agentIndex = 0
    agentMoving = True

#Agent Movement

def move_agent():
    global agentIndex, agentMoving, startPos, agentPos

    if agentIndex >= len(agentPath):
        agentMoving = False
        return

    nextCell = agentPath[agentIndex]

    # Dynamic spawn
    if dynamicMode:
        if random.random() < spawnProbability:
            row = random.randint(0,rows-1)
            col = random.randint(0,cols-1)
            if grid[row][col] == EMPTY:
                grid[row][col] = WALL
                if (row,col) in agentPath[agentIndex:]:
                    agentMoving = False
                    startPos = agentPos
                    clear_search()
                    init_search()
                    global algorithmRunning
                    algorithmRunning = True


    if grid[nextCell[0]][nextCell[1]]==WALL:
        agentMoving=False
        startPos = agentPos
        init_search()
        algorithmRunning=True
        return

    if agentPos:
        if agentPos!=startPos and agentPos!=goalPos:
            grid[agentPos[0]][agentPos[1]]=PATH

    grid[nextCell[0]][nextCell[1]]=AGENT
    agentPos = nextCell
    agentIndex+=1

# Neighbours

def get_neighbors(node):
    r,c=node
    directions=[(1,0),(-1,0),(0,1),(0,-1)]
    result=[]
    for dr,dc in directions:
        nr,nc=r+dr,c+dc
        if 0<=nr<rows and 0<=nc<cols:
            result.append((nr,nc))
    return result

# Main Loop

clock = pygame.time.Clock()
running=True

while running:
    clock.tick(30)
    screen.fill(WHITE)

    pygame.draw.rect(screen,(240,240,240),(600,0,PANEL_WIDTH,WINDOW_HEIGHT))

    for btn in [manhattanBtn,euclideanBtn,astarBtn,gbfsBtn,
                generateBtn,startBtn,pauseBtn,stopBtn, dynamicBtn]:
        btn.draw()

    # Draw Grid
    for r in range(rows):
        for c in range(cols):
            color=WHITE
            if grid[r][c]==WALL: color=GRAY
            elif grid[r][c]==START: color=GREEN
            elif grid[r][c]==GOAL: color=RED
            elif grid[r][c]==PATH: color=BLUE
            elif grid[r][c]==VISITED: color=PURPLE
            elif grid[r][c]==FRONTIER: color=YELLOW
            elif grid[r][c]==AGENT: color=(0,150,0)

            rect=pygame.Rect(c*cellSize,r*cellSize,cellSize,cellSize)
            pygame.draw.rect(screen,color,rect)
            pygame.draw.rect(screen,BLACK,rect,1)

    #Metrics
    
    font = pygame.font.SysFont(None,24)
    screen.blit(font.render(f"Nodes Visited: {nodesVisited}",True,BLACK),(620,550))
    screen.blit(font.render(f"Path Cost: {pathCost}",True,BLACK),(620,580))
    screen.blit(font.render(f"Time (ms): {executionTime}",True,BLACK),(620,610))

    for event in pygame.event.get():
        if event.type==pygame.QUIT:
            running=False

        if event.type==pygame.MOUSEBUTTONDOWN:
            global dynamicMode
            pos=pygame.mouse.get_pos()

            if manhattanBtn.clicked(pos):
                heuristicType="manhattan"
                manhattanBtn.selected=True
                euclideanBtn.selected=False
            elif euclideanBtn.clicked(pos):
                heuristicType="euclidean"
                euclideanBtn.selected=True
                manhattanBtn.selected=False
            elif astarBtn.clicked(pos):
                algorithmType="astar"
                astarBtn.selected=True
                gbfsBtn.selected=False
            elif gbfsBtn.clicked(pos):
                algorithmType="gbfs"
                gbfsBtn.selected=True
                astarBtn.selected=False
            elif generateBtn.clicked(pos):
                generate_random_map()
            elif startBtn.clicked(pos):
                if startPos and goalPos:
                    clear_search()
                    agentPos=startPos
                    init_search()
                    algorithmRunning=True
                    algorithmPaused=False
            elif pauseBtn.clicked(pos):
                algorithmPaused=not algorithmPaused
            elif stopBtn.clicked(pos):
                algorithmRunning=False
                agentMoving=False
                clear_search()
            elif dynamicBtn.clicked(pos):
                dynamicMode = not dynamicMode
                if dynamicMode:
                    dynamicBtn.text = 'Dynamic: ON'
                    dynamicBtn.selected = True
                else:
                    dynamicBtn.text = 'Dynamic: OFF'
                    dynamicBtn.selected = False

            x,y=pos
            if x<GRID_WIDTH and y<rows*cellSize:
                r=y//cellSize
                c=x//cellSize
                if mode=="wall":
                    grid[r][c]=WALL if grid[r][c]==EMPTY else EMPTY
                elif mode=="start":
                    if startPos:
                        prevRow, prevCol=startPos
                        grid[prevRow][prevCol]=EMPTY
                    startPos=(r,c)
                    grid[r][c]=START
                elif mode=="goal":
                    if goalPos:
                        prevRow, prevCol=goalPos
                        grid[prevRow][prevCol]=EMPTY
                    goalPos=(r,c)
                    grid[r][c]=GOAL

        if event.type==pygame.KEYDOWN:
            if event.key==pygame.K_s: mode="start"
            if event.key==pygame.K_g: mode="goal"
            if event.key==pygame.K_w: mode="wall"

    if algorithmRunning and not algorithmPaused:
        if not search_step():
            algorithmRunning=False

    if agentMoving:
        move_agent()

    pygame.display.flip()

pygame.quit()
```
