```python
import pygame
import random
import heapq

pygame.init()

# =======================
# CONFIG
# =======================
WIDTH, HEIGHT = 900, 700
ROWS, COLS = 25, 25
CELL_SIZE = 20
GRID_OFFSET_Y = 120

screen = pygame.display.set_mode((WIDTH, HEIGHT))
pygame.display.set_caption("A* Visualizer - Clean Version")

font = pygame.font.SysFont("arial", 18)

# =======================
# COLORS
# =======================
WHITE = (255,255,255)
BLACK = (0,0,0)
GRAY = (200,200,200)
DARK_GRAY = (150,150,150)

START_COLOR = (0,255,0)
GOAL_COLOR = (255,0,0)
WALL_COLOR = (0,0,0)
FRONTIER_COLOR = (0,0,255)
EXPLORED_COLOR = (255,165,0)
PATH_COLOR = (255,255,0)

BUTTON_COLOR = (100,100,100)
BUTTON_HOVER = (150,150,150)
BUTTON_ACTIVE = (50,200,50)

# =======================
# GRID CLASS
# =======================
class Grid:
    def __init__(self):
        self.grid = [[0 for _ in range(COLS)] for _ in range(ROWS)]
        self.start = None
        self.goal = None

    def reset(self):
        self.grid = [[0 for _ in range(COLS)] for _ in range(ROWS)]
        self.start = None
        self.goal = None

    def generate_walls(self, density):
        for r in range(ROWS):
            for c in range(COLS):
                if random.random() < density:
                    self.grid[r][c] = 1
                else:
                    self.grid[r][c] = 0

    def draw(self):
        for r in range(ROWS):
            for c in range(COLS):
                x = c * CELL_SIZE
                y = r * CELL_SIZE + GRID_OFFSET_Y
                rect = pygame.Rect(x, y, CELL_SIZE, CELL_SIZE)

                if self.grid[r][c] == 1:
                    pygame.draw.rect(screen, WALL_COLOR, rect)
                else:
                    pygame.draw.rect(screen, WHITE, rect)

                pygame.draw.rect(screen, GRAY, rect, 1)

        if self.start:
            self.draw_cell(self.start, START_COLOR)
        if self.goal:
            self.draw_cell(self.goal, GOAL_COLOR)

    def draw_cell(self, pos, color):
        r,c = pos
        rect = pygame.Rect(c*CELL_SIZE, r*CELL_SIZE+GRID_OFFSET_Y, CELL_SIZE, CELL_SIZE)
        pygame.draw.rect(screen, color, rect)

# =======================
# BUTTON CLASS
# =======================
class Button:
    def __init__(self, text, x, y, w, h):
        self.text = text
        self.rect = pygame.Rect(x,y,w,h)
        self.active = False

    def draw(self):
        mouse = pygame.mouse.get_pos()
        color = BUTTON_COLOR

        if self.rect.collidepoint(mouse):
            color = BUTTON_HOVER
        if self.active:
            color = BUTTON_ACTIVE

        pygame.draw.rect(screen, color, self.rect)
        pygame.draw.rect(screen, BLACK, self.rect,2)

        text_surface = font.render(self.text,True,BLACK)
        screen.blit(text_surface,(self.rect.x+10,self.rect.y+8))

    def clicked(self, event):
        if event.type == pygame.MOUSEBUTTONDOWN:
            if self.rect.collidepoint(event.pos):
                return True
        return False

# =======================
# A* CLASS
# =======================
class AStar:
    def __init__(self, grid_obj):
        self.grid_obj = grid_obj
        self.open_set = []
        self.came_from = {}
        self.g_score = {}
        self.f_score = {}
        self.frontier = set()
        self.explored = set()
        self.path = []
        self.running = False
        self.finished = False

    def heuristic(self,a,b):
        return abs(a[0]-b[0]) + abs(a[1]-b[1])

    def start_search(self):
        if not self.grid_obj.start or not self.grid_obj.goal:
            return

        self.open_set = []
        self.came_from = {}
        self.g_score = {}
        self.f_score = {}
        self.frontier = set()
        self.explored = set()
        self.path = []
        self.finished = False

        start = self.grid_obj.start
        self.g_score[start] = 0
        self.f_score[start] = self.heuristic(start,self.grid_obj.goal)
        heapq.heappush(self.open_set,(self.f_score[start],start))
        self.frontier.add(start)
        self.running = True

    def step(self):
        if not self.open_set:
            self.running = False
            self.finished = True
            return

        _,current = heapq.heappop(self.open_set)

        if current == self.grid_obj.goal:
            self.reconstruct_path(current)
            self.running = False
            self.finished = True
            return

        self.frontier.discard(current)
        self.explored.add(current)

        r,c = current
        neighbors = [(r+1,c),(r-1,c),(r,c+1),(r,c-1)]

        for nr,nc in neighbors:
            if 0<=nr<ROWS and 0<=nc<COLS:
                if self.grid_obj.grid[nr][nc] == 1:
                    continue

                tentative_g = self.g_score[current] + 1
                neighbor = (nr,nc)

                if neighbor not in self.g_score or tentative_g < self.g_score[neighbor]:
                    self.came_from[neighbor] = current
                    self.g_score[neighbor] = tentative_g
                    self.f_score[neighbor] = tentative_g + self.heuristic(neighbor,self.grid_obj.goal)
                    heapq.heappush(self.open_set,(self.f_score[neighbor],neighbor))
                    self.frontier.add(neighbor)

    def reconstruct_path(self,current):
        while current in self.came_from:
            self.path.append(current)
            current = self.came_from[current]

    def draw(self):
        for node in self.explored:
            if node != self.grid_obj.start and node != self.grid_obj.goal:
                self.grid_obj.draw_cell(node,EXPLORED_COLOR)

        for node in self.frontier:
            if node != self.grid_obj.start and node != self.grid_obj.goal:
                self.grid_obj.draw_cell(node,FRONTIER_COLOR)

        for node in self.path:
            if node != self.grid_obj.start and node != self.grid_obj.goal:
                self.grid_obj.draw_cell(node,PATH_COLOR)

# =======================
# INIT
# =======================
grid = Grid()
astar = AStar(grid)

run_button = Button("Run A*",20,20,120,40)
reset_button = Button("Reset",160,20,120,40)

density_buttons = [
    Button("10%",320,20,80,40),
    Button("20%",410,20,80,40),
    Button("30%",500,20,80,40),
    Button("40%",590,20,80,40),
]

selected_density = 0.2

clock = pygame.time.Clock()
running = True

# =======================
# MAIN LOOP
# =======================
while running:
    clock.tick(60)
    screen.fill(WHITE)

    for event in pygame.event.get():
        if event.type == pygame.QUIT:
            running = False

        if run_button.clicked(event):
            astar.start_search()

        if reset_button.clicked(event):
            grid.reset()
            astar = AStar(grid)

        for i,btn in enumerate(density_buttons):
            if btn.clicked(event):
                for b in density_buttons:
                    b.active = False
                btn.active = True
                selected_density = (i+1)*0.1
                grid.generate_walls(selected_density)

        if event.type == pygame.MOUSEBUTTONDOWN:
            x,y = event.pos
            if y > GRID_OFFSET_Y:
                col = x // CELL_SIZE
                row = (y - GRID_OFFSET_Y)//CELL_SIZE

                if not grid.start:
                    grid.start = (row,col)
                elif not grid.goal:
                    grid.goal = (row,col)
                else:
                    grid.start = (row,col)
                    grid.goal = None
                    astar = AStar(grid)

    if astar.running:
        astar.step()

    run_button.draw()
    reset_button.draw()
    for btn in density_buttons:
        btn.draw()

    grid.draw()
    astar.draw()

    pygame.display.flip()

pygame.quit()
```
